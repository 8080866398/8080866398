Image to Pencil sketch with python

```python
from google.colab.patches import cv2_imshow
import cv2
```

```python
import cv2
```

```python
img = cv2.imread("p.jpg")
```

```python
cv2_imshow(img)
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_img)
```

```python
inverted_gray_img=cv2.bitwise_not(gray_img)
```

```python
cv2_imshow(inverted_gray_img)
```

```python
blurred_img=cv2.GaussianBlur(inverted_gray_img,(21,21),sigmaX=0,sigmaY=0)
```

```python
cv2_imshow(blurred_img)
```

```python
def dodgev2(x,y):
    return cv2.divide(x, 255-y, scale=255)
```

```python
final_img= dodgev2(gray_img, blurred_img)
```

```python
cv2_imshow(final_img)
```
